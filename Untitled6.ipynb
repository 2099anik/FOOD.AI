```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os
from PIL import Image
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.python.keras.layers import Dense, Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, CSVLogger
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    rescale=1./255,          # Normalize pixel values to be between 0 and 1
    rotation_range=20,       # Randomly rotate images by up to 20 degrees
    width_shift_range=0.2,   # Randomly shift images horizontally by up to 20% of the width
    height_shift_range=0.2,  # Randomly shift images vertically by up to 20% of the height
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=False,
    fill_mode='nearest'
)

batchSize = 32
dataGenerator = datagen.flow_from_directory(
    data_dir,
    target_size=(200, 200),   # Resize images to 200x200 pixels
    batch_size=batchSize,
    class_mode='categorical'
)

augmentedImages, _ = next(dataGenerator)
```

```python
imageHeight,imageWidth=200,200
batchSize=32
```

```python
trainDataset = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=35,
  image_size=(imageHeight, imageWidth),
  batch_size=batchSize)
```

```python
validationDataset = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=35,
  image_size=(imageHeight, imageWidth),
  batch_size=batchSize)
```

```python
classNames = trainDataset.class_names

plt.figure(figsize=(15, 15))
for images, labels in trainDataset.take(1):
    for i in range(min(80, batchSize)):
        ax = plt.subplot(10, 8, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(classNames[labels[i]])
        plt.axis("off")
plt.show()
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten, Dense
import tensorflow as tf

# Load Pretrained ResNet50
pretrainedModel = tf.keras.applications.ResNet50(include_top=False,
                   input_shape=(200,200,3),
                   pooling='avg',
                   weights='imagenet')

# Freeze the layers
for layer in pretrainedModel.layers:
    layer.trainable = False

# Add custom layers on top
x = Flatten()(pretrainedModel.output)
x = Dense(1024, activation='relu')(x)
x = Dense(512, activation='relu')(x)
output = Dense(80, activation='softmax')(x)

# Create final model
resnetModel = Model(inputs=pretrainedModel.input, outputs=output)

# Compile the model
resnetModel.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                     loss='sparse_categorical_crossentropy',
                     metrics=['accuracy'])

resnetModel.summary()
```

```python
resnetModel.compile(optimizer=Adam(learning_rate=0.0001),
                     loss='sparse_categorical_crossentropy',
                     metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=4)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=2)
csv_logger = CSVLogger('ResNet50_Training_History.csv')
epochs=10
history = resnetModel.fit(
  trainDataset,
  validation_data=validationDataset,
  epochs=epochs,
  callbacks=[early_stopping, reduce_lr])
```

```python
import numpy as np
import tensorflow as tf
import cv2
from PIL import Image
from googleapiclient.discovery import build
from youtube_transcript_api import YouTubeTranscriptApi
import openai
from google.colab import files
import io
from IPython.display import display, HTML
from tensorflow.keras.models import load_model

# Load trained model (Upload resnet_model.h5 manually)
model_path = "resnet_model.h5"  # Update path if using Google Drive
model = load_model(model_path)

# YouTube API Key (Replace with your own API Key)

# OpenAI API Key (Replace with your own API Key)
OPENAI_API_KEY = "YOUR_OPENAI_API_KEY"
openai.api_key = OPENAI_API_KEY
# Retrieve class names from trainDataset
data_dir = "/content/drive/MyDrive/Food Classification"
imageHeight,imageWidth=200,200
batchSize=32
trainDataset = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=35,
  image_size=(imageHeight, imageWidth),
  batch_size=batchSize)
classNames = trainDataset.class_names
# Function to preprocess image
# Function to preprocess image (Fix input size)
def preprocess_image(image):
    image = np.array(image)
    if len(image.shape) == 2: # check if image is grayscale
      image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    elif image.shape[2] == 4: #check if image has alpha channel
      image = image[:,:,:3]
    image = cv2.resize(image, (200, 200))  # Match the model's expected input size
    image = image / 255.0  # Normalize
    image = np.expand_dims(image, axis=0)  # Add batch dimension
    return image


# Function to predict dish name
def predict_dish(image):
    processed_image = preprocess_image(image)
    prediction = model.predict(processed_image)
    # Use classNames here instead of the incorrect classes
    predicted_class = classNames[np.argmax(prediction)]
    return predicted_class

# Function to fetch YouTube video
def get_youtube_video(query):
    youtube = build("youtube", "v3", developerKey=YOUTUBE_API_KEY)
    request = youtube.search().list(q=query + " recipe", part="snippet", maxResults=1, type="video")
    response = request.execute()
    video_id = response["items"][0]["id"]["videoId"]
    video_url = f"https://www.youtube.com/watch?v={video_id}"
    return video_url, video_id

# Function to get YouTube transcript
def get_video_transcript(video_id):
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        transcript_text = " ".join([entry["text"] for entry in transcript])
        return transcript_text
    except:
        return "Transcript not available."

# Function to summarize transcript
def summarize_text(text):
    prompt = f"Summarize the following YouTube transcript: {text[:5000]}"  # Limit to avoid API limits
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}]
    )
    return response["choices"][0]["message"]["content"]

# Function to upload and process image in Colab
def upload_and_process():
    uploaded = files.upload()
    for filename in uploaded.keys():
        image = Image.open(io.BytesIO(uploaded[filename]))
        display(image)  # Show uploaded image

        # Predict dish
        predicted_dish = predict_dish(image)
        print(f"\nPredicted Dish: {predicted_dish}")

        # Fetch YouTube video
        video_url, video_id = get_youtube_video(predicted_dish)
        print(f"\nYouTube Recipe Video: {video_url}")
        display(HTML(f'<iframe width="560" height="315" src="https://www.youtube.com/embed/{video_id}" frameborder="0" allowfullscreen></iframe>'))

        # Get transcript and summarize
        transcript = get_video_transcript(video_id)
        if transcript != "Transcript not available.":
            summary = summarize_text(transcript)
            print("\nRecipe Summary:")
            print(summary)
        else:
            print("\nTranscript not available for this video.")

# Run the image upload function
upload_and_process()
```
